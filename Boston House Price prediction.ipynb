```python
#Independent and dependent feature

X = boston.iloc[:,:-1]
y = boston.iloc[:,-1]
```

```python
X.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30, random_state= 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl', 'wb'))
```

```python
X_train
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
reg_pred
```

# Assumption

```python
plt.scatter(y_test, reg_pred)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind= 'kde')
```

```python
#uniform distribution
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print("Mean Squared Error - ",mean_squared_error(y_test, reg_pred))
print("Mean Absolute Error - ", mean_absolute_error(y_test, reg_pred))
print("Root Mean Square Error - ",np.sqrt(mean_squared_error(y_test, reg_pred)))
```

# R Square and adjusted R sqaured

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

# Adjusted R2 = 1 – [(1-R2)*(n-1)/(n-k-1)]

```python
1 - (1-score) * (len(y_test)-1)/ (len(y_test)-X_test.shape[1]-1)
```
